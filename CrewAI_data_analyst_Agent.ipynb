```python
import pandas as pd
from crewai import Agent, Task, Crew
from crewai_tools import FileReadTool
```

```python

df=pd.read_csv('/content/crop_yield_prediction.csv')
df.head()
```

```python
# Define tools
csv_tool = FileReadTool(file_path='/content/crop_yield_prediction.csv')

# Define the Data Analyst agent
data_analyst = Agent(
    role='Data Analyst',
    goal='Analyze the dataset to provide actionable insights.',
    verbose=True,
    memory=True,
    backstory=(
        "You are an expert in data analysis, proficient in identifying trends, patterns, and correlations."
    ),
    tools=[csv_tool],
    allow_delegation=True
)

# Analysis task to analyze the dataset
analysis_task = Task(
    description=(
        "Perform an exploratory data analysis on the dataset. Identify key features, detect missing values, and summarize correlations."
    ),
    expected_output='A detailed summary of dataset analysis, including missing values, correlations, and key insights.',
    tools=[csv_tool],
    agent=data_analyst
)

# Create the crew
data_crew = Crew(
    agents=[data_analyst],
    tasks=[analysis_task],
    process="sequential"
)

# Simulate execution
if __name__ == "__main__":
    # Trigger the crew execution
    results = data_crew.kickoff()

    # Print the final results
    print(results)
```
